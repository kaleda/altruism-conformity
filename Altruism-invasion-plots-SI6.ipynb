This is the code used for plots in Electronic Supplementary Material 6 of "Oblique Transmission, Conformity, and Preference in the Evolution of Altruism."

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_context('paper', font_scale=1.3) 
```

```python
# Inequality 10
def f(β, γ, ρ, b, c, m_v, m_o):
    return b * (ρ * (1 - γ + (m_v * β)) + (1 - ρ) * c * (1 - γ + (m_o * β)))
```

```python
c = [0.2, 1, 1.2, 1.5] 
b = [0.8, 0.9, 1]
ρ = np.linspace(0, 1, 51)
β = np.linspace(0.1, 2.5) 
```

Additive Model I

```python
# Additive Model I
γ = float(input("Enter γ: "))
m_v1 = 1/2
m_o1 = 0

fig, axes = plt.subplots(3, 4, figsize=(16, 12), sharex=True, sharey=True)
fig.suptitle(r'Additive Model I (Parent-to-Offspring Altruism), $\gamma = $'+str(γ), 
             x=0.5, y=0.95, horizontalalignment='center', verticalalignment='top', fontsize = 18)
axes[1,0].yaxis.set_label_coords(-0.18,0.5)
axes[1,0].set_ylabel(r'Vertical transmission rate, $\rho$', fontsize = 15) 
fig.text(0.51, 0.07, r'Fitness benefit, $\beta$', horizontalalignment='center', 
         verticalalignment='bottom', fontsize = 15)

for i in range(3):
    for j in range(4):
        b_, c_ = b[i], c[j]
        ax = axes[i,j]
        β_, ρ_ = np.meshgrid(β, ρ)
        z = f(β_, γ, ρ_, b_, c_, m_v1, m_o1)
        ax.contourf(β_, ρ_, z, levels=[0, 1], colors=['k', 'w'])
        
        if j == 3: #Label b values 
            if i == 2:
                ax.yaxis.set_label_coords(1.24,0.5)
            else:
                ax.yaxis.set_label_coords(1.3,0.5)
            ax.set_ylabel(r'$b = $'+str(b_), horizontalalignment='right', rotation=0)        
        if i == 0: #Label c values
            ax.title.set_position([0.5, 1.045])
            ax.title.set_text(r'$c = $'+str(c_))              
        
        #Label the light and dark regions of the plots
        if γ == 0.01:
            ax.text(1.7, 0.89, 'invasion')
            if j == 0 or (j == 1 and i != 2) or (j == 2 and i == 0):
                ax.text(0.28, 0.07, 'no invasion', fontdict = dict(color = 'w'))
        elif γ == 0.1:
            ax.text(1.7, 0.89, 'invasion')
            if j == 0 or j == 1 or (j == 2 and i == 0):
                ax.text(0.28, 0.07, 'no invasion', fontdict = dict(color = 'w'))
                ax.text(1.7, 0.89, 'invasion')
            if j == 2 and i == 1:
                ax.text(0.15, 0.04, 'no invasion', fontdict = dict(color = 'w'))
        elif γ == 0.3:
            ax.text(1.7, 0.89, 'invasion')
            if j != 3 or i != 2:
                ax.text(0.28, 0.07, 'no invasion', fontdict = dict(color = 'w'))
        elif γ == 0.7:
            ax.text(0.28, 0.07, 'no invasion', fontdict = dict(color = 'w'))
            if i == 1:
                ax.text(1.83, 0.92, 'invasion')
            elif i == 2:
                ax.text(1.8, 0.89, 'invasion')
        elif γ == 0.99:
            ax.text(0.28, 0.07, 'no invasion', fontdict = dict(color = 'w'))
            
#plt.savefig('dec9_mI_g01.pdf') 
```
